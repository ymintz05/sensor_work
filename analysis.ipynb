```python
import numpy as np
import matplotlib.pyplot as plt

resp_sample = "MAINvernier_resp_20260227_105145.npz"
```

```python

with np.load(resp_sample, allow_pickle=True) as d:
    resp = d["data"]                
    ts = d["timestamps"]            
    srate = float(d["srate_hz"])     
    peak = d["peaktrough"]          
    status = d["breath_status"]      

t = ts - ts[0]  

colors = {"inh": "blue", "exh": "red"}

plt.figure()
plt.plot(t, resp, color="black")

for i in range(len(status)-1):
    t0 = status[i][0] - ts[0]
    t1 = status[i+1][0] - ts[0]
    phase = status[i][1]

    if phase in colors:
        plt.axvspan(t0, t1, color=colors[phase], alpha=0.15)

plt.show()
```
